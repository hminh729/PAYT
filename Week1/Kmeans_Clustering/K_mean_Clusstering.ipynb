```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans

img = mpimg.imread('/content/drive/MyDrive/PAYT/Week_1/K-means_Clusstering/Data/man.jpg')
plt.imshow(img)
imgplot = plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
from skimage.transform import rescale
img = rescale(img, 0.5, channel_axis=2)  # Scale ảnh còn 1 nửa
```

```python
plt.imshow(img)
plt.axis('off')
```

```python
X = img.reshape((img.shape[0]*img.shape[1], img.shape[2])) #chuyển từ ảnh từ dạng 3D về dạng 2D
```

```python
for K in [2, 5, 10, 15, 20]:
    kmeans = KMeans(n_clusters=K).fit(X)
    label = kmeans.predict(X)

    img4 = np.zeros_like(X)
    # Thay mỗi pixel có màu tương ứng với từng tâm cụm
    for k in range(K):
        img4[label == k] = kmeans.cluster_centers_[k]
    # Chuyển ảnh từ định dạng 2D về 3D
    img5 = img4.reshape((img.shape[0], img.shape[1], img.shape[2]))
    # Scale ảnh về ban đầu
    img5 = rescale(img5, 2.0, channel_axis=2, order=1)
    plt.imshow(img5, interpolation='nearest')
    plt.axis('off')
    plt.show()
```

```python
# Chọn k = 7
for K in [7]:
    kmeans = KMeans(n_clusters=K).fit(X)
    label = kmeans.predict(X)

    img4 = np.zeros_like(X)
    # Thay mỗi pixel có màu tương ứng với từng tâm cụm
    for k in range(K):
        img4[label == k] = kmeans.cluster_centers_[k]
    # Chuyển ảnh từ định dạng 2D về 3D
    img5 = img4.reshape((img.shape[0], img.shape[1], img.shape[2]))
    # Scale ảnh về ban đầu
    img5 = rescale(img5, 2.0, channel_axis=2, order=1)
    plt.imshow(img5, interpolation='nearest')
    plt.axis('off')
    plt.show()
```
